```python
import numpy as np
import pandas as pd
import tensorflow as tf
import keras 
from keras.models import Sequential
from keras.layers import Dropout,Activation,Dense
from keras.optimizers import serialize
from tensorflow.keras.utils import to_categorical
```

# **Data Loading and Train Test split**

```python
D2 = pd.read_csv("/content/D2.csv")
D2 = pd.DataFrame(D2)
D2.tail()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
X = pd.DataFrame(D2.drop(["class"],axis = 1)).values
y = pd.DataFrame(pd.DataFrame(D2["class"]).values.reshape(-1,1))
y = np.array(pd.get_dummies(y))
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 66302077)
y_test


print(y_test.shape)
```

# **Model Creation**

```python
RNN = Sequential()
RNN.add(Dense(100,activation = "sigmoid",input_dim = 180))
RNN.add(Dropout(0.6))
RNN.add(Dense(80,activation = "sigmoid"))
RNN.add(Dropout(0.6))
RNN.add(Dense(50,activation = "sigmoid"))
RNN.add(Dropout(0.5))
RNN.add(Dense(20,activation = "sigmoid"))
RNN.add(Dropout(0.5))
RNN.add(Dense(3,activation = "softmax"))
```

# **Model compile**

```python
RNN.compile(loss = "categorical_crossentropy",metrics = ["accuracy"])
```

# **Model fitting**

```python
RNN.fit(X_train,y_train,batch_size = 64,epochs = 500)
```

# **Model Evaluation**

```python
score = RNN.evaluate(X_test,y_test,batch_size = 64)
```

```python
y_pred = RNN.predict(X_test)
np.round(y_pred,5)
```

```python
y_original_pred = np.argmax(y_pred,axis = 1)
```
